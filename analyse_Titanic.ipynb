## Connecting to GitHub in Google Colab

There are two main ways to connect your Colab environment with GitHub:

1.  **Through the Colab UI (for notebooks):**
    *   **Saving a copy to GitHub:** Go to `File > Save a copy to GitHub`. You'll be prompted to authenticate with your GitHub account.
    *   **Opening a notebook from GitHub:** Go to `File > Open notebook` and select the 'GitHub' tab. You can then paste a GitHub URL or browse your repositories.

2.  **Using Git commands directly in your notebook (for repositories and version control):**
    You can use `!git` commands in your code cells to clone repositories, make commits, and push changes. You'll typically need to set up your user name and email, and often use a Personal Access Token (PAT) for authentication, especially when pushing changes.
